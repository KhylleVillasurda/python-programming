```python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import torch

##Verify Dependencies Installation and Versions
print(scipy.__version__, sklearn.__version__, matplotlib.__version__, torch.__version__)
```

```python
X, y = make_blobs(n_samples=600, centers=2, cluster_std=1, random_state=42)

data = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
data['TrueLabel'] = y
data.head()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data["Feature1"], data["Feature2"], c=labels, cmap="viridis", s=30)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', marker='.', s=800, label='Centroids')
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
